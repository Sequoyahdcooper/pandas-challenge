```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"]) 
school_data_complete
```

```python
Total_Schools = school_data_complete["school_name"].nunique() 
print('Total Schools =  ', Total_Schools)

Total_Students = school_data_complete["student_name"].nunique()
print('Total Students =  ', Total_Students)

Total_Budget = school_data_complete['budget'].sum()
print('Total Budget =  ', Total_Budget)

Avg_math_score = round(school_data_complete["math_score"].mean(),1)
print('Avg math score =  ', Avg_math_score)

Avg_reading_score = round(school_data_complete["reading_score"].mean(),1)
print('Avg reading score =  ', Avg_reading_score) 

#Finding the amount of math scores grater than 70
Math_score_filter = school_data_complete['math_score'] > 70 
#print(Math_score_filter)


filtered_math = school_data_complete['math_score'][Math_score_filter]
#print(filtered_math) 

Count_mScore_above_seventy = len(filtered_math.index) 
print('# of students with math scores above 70 =  ',Count_mScore_above_seventy)

Math_score_percent_greater_seventy = (Count_mScore_above_seventy/Total_Students)*100
print('% of students with math score above 70 =  ', Math_score_percent_greater_seventy) 

#Finding the amount of reading scores grater than 70
Reading_score_filter = school_data_complete['reading_score'] > 70 
#print(Reading_score_filter)


filtered_reading = school_data_complete['reading_score'][Reading_score_filter]
#print(filtered_reading) 

Count_rScore_above_seventy = len(filtered_reading.index) 
print('# of students with reading scores above 70 =  ',Count_rScore_above_seventy)

Reading_score_percent_greater_seventy = (Count_rScore_above_seventy/Total_Students)*100
print('% of students with reading scores above 70 =  ', Reading_score_percent_greater_seventy)

#Overall Passing Rate
overall_passing_rate = ((Math_score_percent_greater_seventy + Reading_score_percent_greater_seventy)/2)
print('The overall passing rate is = ', overall_passing_rate)
```
